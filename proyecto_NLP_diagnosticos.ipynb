### Tokenización de la columna criterios

```python
texto = df['Criterios']
```

```python
# Obtengo el corpus de las palabras
corpus = list()
for diagnostico in texto:
    palabras = diagnostico.split(" ")
    for palabra in palabras:
        corpus.append(palabra)
```

```python
#Tokenizo todas las palabras del corpus
tokens= list()
for palabra in corpus:
    token=nltk.word_tokenize(palabra,"spanish")
    for elemento in token:
        tokens.append(elemento)
```

```python
# Limpio tokens de stopwords y caracteres especiales
tokens_limpios = list()
for token in tokens:
    if token not in palabras_funcionales:
        if len(token) > 2:
            if token != '...':       
                tokens_limpios.append(token)
print(len(tokens))
print(len(tokens_limpios))
```

### Creo la matriz TF-IDF

```python
count = CountVectorizer(binary= True)

bag = count.fit_transform(df['Criterios'])
```

```python
bag1 = bag.toarray()
```

```python
bag1
```

```python
palabras = pd.DataFrame(bag1, columns=count.get_feature_names_out())
palabras
```

```python
for palabra in palabras:
    if palabra in palabras_funcionales or len(palabra) <= 2 or palabra == '...':
        palabras = palabras.drop(palabra,axis=1)
```

```python
for palabra in palabras:
    if palabras[palabra].sum() > 50:
        palabras = palabras.drop(palabra,axis=1)
```

```python
tfidf = TfidfTransformer()

TF_IDF_sk=pd.DataFrame(tfidf.fit_transform(palabras).toarray(),columns=palabras.columns)
TF_IDF_sk
```

### Prueba PCA + K-Means

```python
pca = PCA(n_components=4) 
pca = pca.fit(TF_IDF_sk)
```

```python
data_transformed=pca.transform(TF_IDF_sk)
```

```python
pca.explained_variance_ratio_
```

```python
reduced_data = pca.transform(TF_IDF_sk)
reduced_data = pd.DataFrame(reduced_data, columns = ['Dimension 1', 'Dimension 2','Dimension 3','Dimension 4'])
```

```python
# Poner grupos suficientes para explicar un 80% de la varianza
kmeans = KMeans(4)
cluster_data=reduced_data[["Dimension 1","Dimension 2",'Dimension 3','Dimension 4']]
kmeans.fit(cluster_data)
reduced_data["clusters"]=kmeans.labels_
```

```python
df['clusters']=reduced_data['clusters']
```
